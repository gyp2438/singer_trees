```python
from coloring_groups import EdgeColoring
```

```python
kappa = EdgeColoring([
    [(2,3),(4,5)],
    [(3,4),(6,7)],
    [(1,2),(3,6)]
]);

kappa.plot()
```

```python
G = kappa.group();

print(G.structure_description())

gens = G.gens()
```

```python
cox_elt = gens[0]*gens[1]*gens[2];

conjugates = set();

for g in G:
    conjugates.add(g*cox_elt*g^(-1))
```

```python
print("Conjugates of g0*g1*g2:")
for c in conjugates:
    print("\t", c)
```

```python
print("Coxeter Elements | Is conjugate to g0*g1*g2");

for elt in kappa.coxeter_elements():
    print("\t", elt, "|", elt in conjugates)
```

```python
absolute_order = kappa.absolute_order();
absolute_order.plot(figsize = 20, label_elements = False)
```

```python
cox_elts = list(conjugates)
intervals = [absolute_order.subposet(absolute_order.order_ideal([elt])) for elt in cox_elts];

print("Coxeter Element | is lattice? | is self-dual? | is isomorphic to intervals[0]?")

for i in range(0, len(intervals)):
    print(cox_elts[i], "|", intervals[i].is_lattice(), "|", intervals[i].is_self_dual(), "|", intervals[i].is_isomorphic(intervals[0]));
```
